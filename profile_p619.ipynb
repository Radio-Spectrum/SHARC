```python
import pandas as pd
# Opcional: gerar um gráfico para visualização
import matplotlib.pyplot as plt
```

```python
# Carregar o arquivo CSV
df = pd.read_csv('profile_stats_gasses.csv')

# Exibir as primeiras linhas do DataFrame
print(df.head())

```

```python
# Ordenar por tempo acumulado para encontrar os gargalos
sorted_df = df.sort_values(by='Cumulative Time', ascending=False)
print(sorted_df.head(10))  # Mostrar as 10 funções que mais consomem tempo


```

```python

top_funcs = sorted_df.head(10)
plt.barh(top_funcs['Function'], top_funcs['Cumulative Time'])
plt.xlabel('Cumulative Time (s)')
plt.ylabel('Function')
plt.title('Top 10 Gargalos de Desempenho')
plt.gca().invert_yaxis()  # Inverter a ordem para a barra mais alta no topo
plt.show()
```
